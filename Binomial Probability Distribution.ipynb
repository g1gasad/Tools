```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')

def factorial(x):
    n = 1
    while x != 0:
        n = n * x
        x = x - 1
    return n

def combination(n, k):
    c = factorial(n) / (factorial(n - k) * factorial(k))
    return c

def binomial_distribution(n, k, p):
    """
        Here, 
        n: The given number of trials
        k: Number of Successes
        p: Probability of success on a given trial
    """
    likelihood = combination(n, k) * p**k * (1 - p)**(n - k)
    return round(likelihood, 4)

binomial_distribution(10, 1, 0.5)

def distrubution_data(n, p):
    points = {}
    values = [binomial_distribution(n, k, p) for k in range(n+1)]
    keys = [k for k in range(n+1)]
    points['Keys'] = keys
    points['Value'] = values
    return points

def distribution_graph(n, p):
    data = distrubution_data(n, p)
    plt.figure(figsize=(10,4))
    sns.barplot(data=data, x='Keys', y='Value',
                edgecolor='black',
                width=0.8,
                color='violet')
    plt.show()

n = int(input("Enter the total number of Trials: "))
p = float(input('Enter the probabilty of success of each trial'))
distribution_graph(n, p)
```
