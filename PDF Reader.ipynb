```python
import pdfplumber
import pandas as pd

# Path to the PDF file
pdf_path = 'Downloads/Divisomnal office address and contact number.pdf'

# Open the PDF file
pdf = pdfplumber.open(pdf_path)

# Get the pages containing the tables (replace [6, 7] with the correct page indices)
table_pages = [pdf.pages[6], pdf.pages[7]]

# Initialize an empty list to store extracted tables
tables = []

# Extract and process tables from each page
for page in table_pages:
    table = page.extract_table()
    if table:
        tables.append(table)

# Close the PDF file
pdf.close()

# Concatenate the extracted tables into a single DataFrame
if tables:
    concatenated_table = pd.concat([pd.DataFrame(table) for table in tables], ignore_index=True)
    # print(concatenated_table)
else:
    print("No tables extracted")

# Note: The above code assumes that the extracted tables have the same structure.
```
